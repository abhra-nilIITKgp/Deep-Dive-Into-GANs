```python
import model
import trainer
import matplotlib.pyplot as plt
import torch
import torchvision
import torchvision.transforms as transforms
```

```python
mnist = torchvision.datasets.MNIST('./MNIST',download=True,transform=transforms.Compose([transforms.ToTensor(),transforms.Normalize((.5,.5,.5),(.5,.5,.5))]))
generator = model.Generator()
discriminator = model.Discriminator()
device = torch.device('cuda:0')
Trainer = trainer.GANTrainer(device,G=generator,D=discriminator,dataset=mnist,batch_size=32,epochs=400,lr=0.0002,model='./GAN.model',images='./Images')
x = int(torch.randint(0,len(mnist),(1,0)).item())
print(x)
sample = mnist[x][0]
sample = torch.unsqueeze(sample,0)
sample.shape
```

```python
#Trainer.load_model()
```

```python
Trainer.train()
```

```python
plt.plot(Trainer.losses_g)
plt.xlabel('Iterations')
plt.ylabel('Generator Loss')
```
